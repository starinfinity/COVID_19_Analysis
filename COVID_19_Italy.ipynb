```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os

from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
```

```python
covid_19_data = pd.read_csv("corona-virus-report/covid_19_clean_complete.csv")
```

```python
covid_19_data.head(6)
```

```python
df = covid_19_data[covid_19_data['Country/Region'] == "Italy"]
df.head(6)
```

```python
confirmed_case = df.drop(['Confirmed', 'Deaths', 'Recovered'], axis=1)
confirmed_case['Count'] = df['Confirmed']
confirmed_case['Type'] = 'Confirmed'
```

```python
confirmed_case.head(6)
```

```python
deaths = df.drop(['Confirmed', 'Deaths', 'Recovered'], axis=1)
deaths['Count'] = df['Deaths']
deaths['Type'] = 'Deaths'
```

```python
deaths.head(6)
```

```python
recovered = df.drop(['Confirmed', 'Deaths', 'Recovered'], axis=1)
recovered['Count'] = df['Recovered']
recovered['Type'] = 'Recovered'
```

```python
recovered.head(6)
```

```python
frames = [confirmed_case, deaths, recovered]
df_EDA = pd.concat(frames, sort=False)
```

```python
df_EDA['Date'] = pd.to_datetime(df_EDA['Date']) 
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
```

```python
x = sns.lineplot(x=df_EDA['Date'], y=df_EDA['Count'], hue=df_EDA['Type'])
rotation = plt.setp(x.get_xticklabels(), rotation=45)
```
